```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

density_data1 = [
    [2, 3, 18521],
    [3, 3, 1136],
    [4, 3, 49],
    [5, 3, 4],
    [6, 3, 0]
]
density_data2 = [
    [3, 3, 1136],
    [3, 4, 201],
    [3, 5, 44],
    [3, 6, 7],
    [3, 7, 1]
]
density_data_columns = ['dim', 'bound', 'ideal_lattice_num']
df1 = pd.DataFrame(density_data1, columns=density_data_columns)
df2 = pd.DataFrame(density_data2, columns=density_data_columns)

def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + 200
                value = '{}'.format(int(p.get_height()))
                ax.text(_x, _y, value, ha="center", color='black') 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height() * 0.5)
                value = '{}'.format(int(p.get_height()))
                ax.text(_x, _y, value, ha="left", color='black')

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)


p = sns.barplot(data=df1, x='dim', y='ideal_lattice_num', palette=colors)

df1['dim'] = df1['dim'] - 2

show_values(p)

sns.lineplot(x='dim', y='ideal_lattice_num', data=df1)

plt.xlabel('Dimension (dim)')
plt.ylabel('Ideal Lattice Number')
plt.title('Ideal Lattice Number (Bound Fixed)')

plt.legend(title='Bound = 3', loc='upper right')

plt.show()
```
